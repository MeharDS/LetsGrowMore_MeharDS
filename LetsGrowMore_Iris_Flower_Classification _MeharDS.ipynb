# Model Building:

```python
final_df=df.drop('species_encoded', axis=1)
```

```python
# Split data into X and y
X=final_df.drop('species', axis=1)
y=final_df['species']
```

```python
#Split data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import  KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
lr = LogisticRegression()
knc= KNeighborsClassifier(n_neighbors=15,algorithm="auto")
rfc=RandomForestClassifier()
```

```python
#creating dictuory for clfs
clfs={
    'lr':LogisticRegression(),
    'knc':KNeighborsClassifier(),
    'rfc':RandomForestClassifier()
    }
```

```python
#Function for train model
def train_classifer(clf, X_train, X_test, y_train, y_test):
    from sklearn.metrics import accuracy_score
    clf.fit(X_train,y_train)
    print(f'{clf} Train Score is : ' , clf.score(X_train, y_train))
    print(f'{clf} Test Score is : ' , clf.score(X_test, y_test))
    y_pred = clf.predict(X_test)
    score = accuracy_score(y_test, y_pred)
    return score
```

```python
scores_list=[]
for name, clf in clfs.items():    
    current_score = train_classifer(clf, X_train, X_test, y_train, y_test)
    scores_list.append(current_score)
```

# Create Perfromance Data Frame:

```python
performance_df = pd.DataFrame({'Algorithm': clfs.keys(), 'Score':scores_list})
```
